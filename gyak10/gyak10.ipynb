# Matplotlib 

A Matplotlib egy alacsony szintű gráfrajzoló könyvtár a Pythonban, amely vizualizációs segédprogramként szolgál. Főként a pyplot nevű almodullal fogunk dolgozni, amelyet plt néven importálunk:

```python
import matplotlib.pyplot as plt
import numpy as np
```

Kezdetként kössünk össze vele két pontot, legyenek ezek a (0, 0) és a (6, 250). Ehhez két tömböt készítünk, az xpoints tömbben soroljuk fel a pontok x, az ypoints tömbben a pontok y koordinátáit.
Ez a két tömb lesz az ábrázoló `plt.plot` függvény két bemenete.

```python
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

Ha a két pontot nem összekötni szeretnénk, hanem markerrel megjelölni őket, akkor a plot függvénynek megadunk egy harmadik paramétert is, pl. ha karikákat szeretnénk markerként használni, akkor ez a paraméter: 'o'

```python
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints, linestyle='dashed',marker='^')
plt.show()
```

Kettőnél tömb pontot is összeköthetünk egyenes szakaszokkal:

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

Az x-koordinátákat tartalmazó tömb deafult értéke egy 0,1,2,... elemű tömb, az utolsó elemét az határozza meg, hogy hány elemű az y-koordinátákat tartalmazó tömb.

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

A karikán kívül sok más markert is használhatunk. A marker és a pontokat összekötő vonal stílusa és színe is megválasztható, pl. piros pontozott vonallal így köthetjük össze a pontok helyére rakott piros pöttyöket (':' = pontozott vonal, 'r' = piros):

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, '^:C1')
plt.show()
```

A marker méretén is válroztathatunk a markersize vagy ms argumentum megadásával:

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

A markerek határvonalának/belsejének a színét is külön beállíthatjuk (határ: markeredgecolor vagy mec, belső rész: markerfacecolor vagy mfc):

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

További kulcsszó-argumentumok: linestyle (ls), linewidth (lw), color (c)

Több grafikon is készíthető egy ábrában:

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])


plt.figure()
plt.plot(x1, y1, x2, y2)
plt.show()

plt.figure()
plt.plot(x1, y1, 'r')
plt.plot(x2, y2,'b')
plt.show()
```

Az x-tengelyt az xlabel(), az y-tengelyt pedig az ylabel() függvény segítségével láthatjuk el felirattal, és az egész ábrának címet is adhatunk a title() függvénnyel:

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Két görbe egy ábrában')
plt.show()
```

A betűtípust is megválaszthatjuk a fontdict paraméter segítségével. 

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.figure(figsize=(20,10))

font = {'family':'serif','color':'blue','size':20}
plt.plot(x1, y1, x2, y2)
plt.xlabel('x', fontsize=20)
plt.ylabel('y')
plt.title('Két görbe egy ábrában', fontdict = font)
plt.show()
```

A cím alapértelmezésben középre kerül, de a loc paraméterrel jobbra vagy balra is csúsztathatjuk.

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

font = {'family':'serif','color':'blue','size':20}
plt.plot(x1, y1, x2, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Két görbe egy ábrában', fontdict = font, loc = 'left')
plt.show()
```

Rácshálót is készíthetünk a grid() függvénnyel:

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

font = {'family':'serif','color':'blue','size':20}
plt.plot(x1, y1, x2, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Két görbe egy ábrában', fontdict = font, loc = 'left')
plt.grid()
plt.show()
```

Ha csak az x-tengely vagy az y-tengely pontjain átmenő rácsvonalakra van szükségünk, akkor a grid() függvénynek axis = 'x' ill. axis = 'y' módon:

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

font = {'family':'serif','color':'blue','size':20}
plt.plot(x1, y1, x2, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Két görbe egy ábrában', fontdict = font, loc = 'left')
plt.grid(axis = 'y')
plt.show()
```

A rácsvonalak színét, stílusát és vastagságát is meghatározhatjuk:

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

font = {'family':'serif','color':'blue','size':20}
plt.plot(x1, y1, x2, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Két görbe egy ábrában', fontdict = font, loc = 'left')
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
plt.figure(figsize=(5,5))
x=np.linspace(0, 2*np.pi,100)
plt.plot(np.sin(x),np.cos(x))
plt.show()
```

```python
def draw_circle(x0,y0,r):
    plt.figure(figsize=(5,5))
    x=np.linspace(0, 2*np.pi,100)
    plt.plot(r*np.sin(x)+x0,r*np.cos(x)+y0)
    plt.show()
```

```python
draw_circle(1,1,0.5)
```

```python
def draw_spiral(a,b):
    plt.figure(figsize=(5,5))
    phi=np.linspace(0, 2*np.pi,100)
    r= a+b*phi
    plt.plot(r*np.sin(x),r*np.cos(x))
    plt.show()
```

```python
draw_spiral(1,0.5)
```

```python
def draw_spiral2(a,k):
    plt.figure(figsize=(5,5))
    r=np.linspace(10e-10,10,1000)
    phi=1/k * np.log(r/a)
    plt.plot(r*np.sin(phi),r*np.cos(phi))
    plt.show()
```

```python
draw_spiral2(0.6,0.15)
```
